```python
import numpy as np
import matplotlib.pyplot as plt

# Generate normal (Gaussian) data
np.random.seed(42)

X1 = np.random.normal(loc=2, scale=0.5, size=(50, 2))
X2 = np.random.normal(loc=6, scale=0.5, size=(50, 2))
X3 = np.random.normal(loc=10, scale=0.5, size=(50, 2))

X = np.vstack((X1, X2, X3))

# Number of clusters
k = 3

# Random initialization of centroids
centroids = X[np.random.choice(len(X), k, replace=False)]

# K-Means algorithm
for i in range(10):
    clusters = [[] for _ in range(k)]
    
    # Assignment step
    for point in X:
        distances = [np.linalg.norm(point - centroid) for centroid in centroids]
        cluster_index = np.argmin(distances)
        clusters[cluster_index].append(point)
    
    # Update step
    centroids = np.array([np.mean(cluster, axis=0) for cluster in clusters])

# Plot clusters
for cluster in clusters:
    cluster = np.array(cluster)
    plt.scatter(cluster[:, 0], cluster[:, 1])

# Plot centroids
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x')

plt.xlabel("X1")
plt.ylabel("X2")
plt.title("K-Means Clustering on Normal Distributed Data")
plt.show()
```
