```python
#count the number of trailing zeros in a binary number
def count(N):
    count=0
    if N==0:
        return 0
    while(N&1)==0:
        count+=1
        N=N>>1
    return count
N=int(input("Enter a number : "))
print("The number of trailing zeros are",count(N))
```

```python
#finds the exponential of x power y
def exponential(x,y):
    if y==0:
        return 1
    else:
        return x*exponential(x,y-1)
x=int(input("Enter a number : "))
y=int(input("Enter a number : "))
print("The Exponential is",exponential(x,y))
```
