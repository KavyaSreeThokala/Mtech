```python
# Open the file in read mode
file_name = input("Enter the file name: ")

tabs = 0
spaces = 0
newlines = 0

with open(file_name, "r") as file:
    content = file.read()
    
    for ch in content:
        if ch == '\t':
            tabs += 1
        elif ch == ' ':
            spaces += 1
        elif ch == '\n':
            newlines += 1

# Display the results
print("Number of tabs:", tabs)
print("Number of spaces:", spaces)
print("Number of newline characters:", newlines)
```

```python
# Input file names
source_file = input("Enter source binary file name: ")
destination_file = input("Enter destination file name: ")

# Open source file in binary read mode
with open(source_file, "rb") as src:
    data = src.read(10)   # Read first 10 bytes

# Open destination file in binary write mode
with open(destination_file, "wb") as dest:
    dest.write(data)      # Write first 10 bytes

print("First 10 bytes copied successfully.")
```
