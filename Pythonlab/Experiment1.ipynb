```python
# Integer variable
age = 20
# Float variable
height = 5.8
# String variable
name = "Kavya"
# Boolean variable
is_student = True
# Literal constants
pi = 3.14
grade = 'A'
message = "Welcome to Python Programming!"
# Displaying the values and their data types
print("Age:", age, "Type:", type(age))
print("Height:", height, "Type:", type(height))
print("Name:", name, "Type:", type(name))
print("Is Student:", is_student, "Type:", type(is_student))
print("Pi Value:", pi, "Type:", type(pi))
print("Grade:", grade, "Type:", type(grade))
print("Message:", message, "Type:", type(message))
```

```python
# Initial string
message = "Hello"
# Using += operator to append another string
message += " World"
# Display the result
print("Updated String:", message)
```

```python
# String variables
str1 = "Hello"
str2 = "Python"
# Addition (Concatenation)
add_result = str1 + " " + str2
# Multiplication (Repetition)
mul_result = str1 * 3
# Display results
print("String Addition Result:", add_result)
print("String Multiplication Result:", mul_result)
```
