```python
# Take number of elements
n = int(input("Enter the number of integers: "))

# Original list
original_list = []

# Input n integers
for i in range(n):
    num = int(input(f"Enter integer {i+1}: "))
    original_list.append(num)

# Create a new list without duplicates
unique_list = []
for num in original_list:
    if num not in unique_list:
        unique_list.append(num)

# Print both lists
print("Original list:", original_list)
print("List without duplicates:", unique_list)
```

```python
# Create a tuple
nums = (10, 20, 30, 40, 50)

# Take input from the user
num = int(input("Enter a number to search: "))

# Check if the number exists in the tuple
if num in nums:
    index = nums.index(num)
    print(f"Number {num} is present at index {index}.")
else:
    print("The number is not present in the tuple.")
```

```python
# Read first sequence
n1 = int(input("Enter number of elements in first sequence: "))
seq1 = []

for i in range(n1):
    seq1.append(int(input(f"Enter element {i+1}: ")))

# Read second sequence
n2 = int(input("Enter number of elements in second sequence: "))
seq2 = []

for i in range(n2):
    seq2.append(int(input(f"Enter element {i+1}: ")))

# Convert lists to sets to remove duplicates
set1 = set(seq1)
set2 = set(seq2)

# Common elements
common_elements = sorted(set1 & set2)

# Symmetric difference (elements in exactly one sequence)
symmetric_diff = sorted(set1 ^ set2)

# Print results
print("Common elements:", common_elements)
print("Elements appearing in exactly one sequence:", symmetric_diff)
```
