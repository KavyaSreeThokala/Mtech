```python
# Take a number as input from the user
num = float(input("Enter a number: "))

# Lambda function to square the number
square = lambda x: x ** 2

# Calculate squared value
squared_value = square(num)

# Print the result
print("Squared value:", squared_value)
```

```python
# Recursive function to count digits
def count_digits(n):
    if n == 0:
        return 0
    else:
        return 1 + count_digits(n // 10)

# Input from the user
num = int(input("Enter a positive integer: "))

# Handle the case when the number is 0
if num == 0:
    print("Number of digits: 1")
else:
    print("Number of digits:", count_digits(num))
```

```python
# Take number of elements
n = int(input("Enter the number of integers: "))

# Initialize an empty list
numbers = []

# Input n integers
for i in range(n):
    num = int(input(f"Enter integer {i+1}: "))
    numbers.append(num)

# Print the list in original order
print("Original list:", numbers)

# Print the list in reverse order without using reverse() or [::-1]
print("Reversed list:", end=" ")
for i in range(n-1, -1, -1):
    print(numbers[i], end=" ")
print()
```
