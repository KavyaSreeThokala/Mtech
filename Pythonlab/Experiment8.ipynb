```python
# Import random module with alias
import random as rnd

# Print 5 random integers between 1 and 100
print("5 random integers between 1 and 100:")
for i in range(5):
    print(rnd.randint(1, 100))
```

```python
# Import pyplot from matplotlib with alias
import matplotlib.pyplot as plt

# Define data
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# Plot the line graph
plt.plot(x, y)

# Add labels and title
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Simple Line Graph")

# Display the graph
plt.show()
```
